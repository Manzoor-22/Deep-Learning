### Step-1 : Loading the libraries

```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.regularizers import l2
```

### Step-2 : Loading the dataset

```python
(train_images, train_labels),(test_images, test_labels) = mnist.load_data()
train_images = train_images.reshape((60000, 28, 28, 1)).astype('float32') / 255
test_images = test_images.reshape((10000, 28, 28, 1)).astype('float32') / 255
```

```python
train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)
```

### Step-3 : Creating the model

```python
model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', kernel_regularizer='l2', input_shape=(28,28,1)),
    layers.MaxPool2D((2,2)),
    layers.Conv2D(64, (3,3), activation='relu', kernel_regularizer='l2'),
    layers.MaxPool2D((2,2)),
    layers.Conv2D(64, (3,3), activation='relu', kernel_regularizer='l2'),
    layers.Flatten(),
    layers.Dense(64, activation='relu', kernel_regularizer = 'l2'),
    layers.Dropout(0.5),
    layers.Dense(10, activation='softmax')
])
```

### Step-4 : Fitting on model

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
gen_model = model.fit(train_images, train_labels, epochs=5, batch_size=64, validation_split=0.2)
```

```python
test_loss, test_acc = model.evaluate(test_images,test_labels)

print(f"Test accuracy = {test_acc}")
```

### Step-5 : Plotting the graph

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(gen_model.history['accuracy'], label = "Training Accuracy")
plt.plot(gen_model.history['val_accuracy'], label = "Validation Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```
